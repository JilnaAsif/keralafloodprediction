```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='FLOODS', y=' ANNUAL RAINFALL', palette='Set2')
plt.title('Annual rainfall vs chance of flood')
plt.xlabel('Flood')
plt.ylabel('Annual Rainfall')
plt.show()
```

* There was flood when annual rainfall is between 3000 and 3500.
* Tere are some outliers in annulal rainfall when flood occured. Since it is a significant data, we can not treat this outlier.
* There is no outliers when there is no flood. so if the annual rainfall is around 2500 there is no chance for flood.
* The annual rainfall influences occurences of flood.

### Annual Rainfall Over Years

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df,x='YEAR',y=' ANNUAL RAINFALL')
plt.title("Annual Rain Fall over Years")
plt.show()
```

* This gragh shows the annual rainfall over years. Maximum rainfall occured in 2018.

```python
df.columns = df.columns.str.strip()
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
monthly_avg_rainfall = df[months].mean()
plt.figure(figsize=(12, 6))
plt.bar(months, monthly_avg_rainfall, color='skyblue', edgecolor='black')
plt.title('Average Monthly Rainfall in Kerala', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Rainfall (mm)', fontsize=14)
plt.show()
```

* Maximum average rainfall occured during the months of jun, july and august. 

### flood years vs non-flood years

```python
flood_counts = df['FLOODS'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(flood_counts, labels=flood_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightgreen'])
plt.title('Flood Years vs Non-Flood Years in Kerala', fontsize=16)
plt.axis('equal')  # Makes the pie chart a perfect circle
plt.show()
```

* From 1900 to 2020, about half of the years experienced floods.

### Monthly Rainfall in Kerala(2018)

```python
data=df[df['YEAR']==2018].squeeze()
month=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
plt.figure(figsize=(12, 6))
plt.bar(month, data[month], color='skyblue', edgecolor='black')
plt.title('Monthly Rainfall in Kerala(2018)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Rainfall (mm)', fontsize=14)
plt.show()
```

* As already seen the maximum rainfall occured in 2018. In 2018 the rain fall increasing from jun and the rainfall was maximum in august.
